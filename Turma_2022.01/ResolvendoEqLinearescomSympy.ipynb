# Resolvendo equações lineares com Sympy

Sympy é uma biblioteca de Python. O primeiro passo é importá-la no presente notebook. Geralmente, usamos a abreviação *sym* para Sympy.  

```python
import sympy as sym
```

O sistema de equações a ser solucionado é 

\begin{align*}
n_1 + n_1 & = 100 \\
0.7 n_1 + 0.2 n_2 & = 30
\end{align*}

O próximo passo é introduzir nomes para as variáveis desconhecidas que aparecem no nosso problema. A função Sympy para isso é *sym.var()* que constrói variáveis simbólicas dada uma lista de nomes de variáveis. 

```python
sym.var(['n1','n2'])
```

As novas variáveis simbólicas são usadas para criar as equações simbólicas. 

A função *sym.Eq()* aceita dois argumentos, cada argumento é uma expressão simbólica do lado esquerdo e do lado direito de uma equação. 

Para esse problema, temos duas equações que devem ser resolvidas simultaneamente, então construímos ambas e estocamos as mesmas dentro de uma lista. 


```python
eqns = [
    sym.Eq(n1 + n2, 100),
    sym.Eq(0.7*n1 + 0.2*n2, 30)
]
print(eqns)
```

O último passo é resolver as equações utilizando *sym.solve()*. 

```python
soln = sym.solve(eqns)
print(soln)
```

Colocando tudo junto, temos um procedimento em **três** etapas para resolver um sistema de equações lineares com Sympy. 

```python
# import sympy
import sympy as sym

# Step 1. Create symbolic variables.
sym.var(['n1','n2'])

# Step 2. Create a list of equations using the symbolic variables
eqns = [
    sym.Eq(n1 + n2, 100),
    sym.Eq(0.7*n1 + 0.2*n2, 30)
]

# Step 3. Solve and display solution
soln = sym.solve(eqns)
print(soln)
```
